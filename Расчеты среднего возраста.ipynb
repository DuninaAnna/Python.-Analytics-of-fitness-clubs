Вам предстоит поработать со списком возрастов посетителей фитнес-клубов age_lst

```python
import numpy as np
```

```python
age_lst = [23, 45, 18, 80, 22, 25, 34, 37, 41, 88, 19, 20, 18, 45, 41, 39, 31, 32, 22, 28, 37]
```

**Задача 1.** Рассчитайте средний возраст по всему списку age_lst

```python
avg_age = round(sum(age_lst)/len(age_lst),0)
print (f'Средний возраст пользователя {avg_age} лет.')
```

**Задача 2.** Рассчитайте средний возраст для посетителей, которые моложе 50 лет

```python
sum_age = 0
cnt_people = 0
for i in age_lst:
    if i < 50:
        sum_age +=i
        cnt_people +=1
    avg_age = round(sum_age/cnt_people,0)
print (f'Средний возраст пользователя {avg_age} лет.')
```

**Задача 3.** Оберните предыдущий код в функцию, которая на вход принимает список и пороговое значение

```python
def middle_age(lst, border):
    sum_age = 0
    cnt_people = 0
    for i in age_lst:
        try:
            if i < border:
                sum_age +=i
                cnt_people +=1
        except TypeError:
            print (f'Значение {i} не является числом')  
    try:    
        avg_age = round(sum_age/cnt_people,0)
        print (f'Средний возраст пользователя {avg_age} лет.')
        return round(avg_age,0)
    except:
         (f'Список должен содержать хотя бы одно значение меньше {border}')
```

```python
middle_age(age_lst,31)
```

**Задача 4.** Рассчитайте среднее значение для всех элементов списка, за исключением тех, которые лежат вне интервала [Q1-1,5*IQR; Q3+1,5*IQR]


```python
q1 = np.percentile(age_lst,25)
q3 = np.percentile(age_lst,75)
iqr = q3-q1
lower_border = q1-1.5*iqr
upper_border = q3+1.5*iqr
sum_age = 0
cnt_people = 0
for i in age_lst:
    if i < upper_border and i > lower_border:
        sum_age +=i
        cnt_people +=1
    avg_age = round(sum_age/cnt_people,0)
print (f'Средний возраст пользователя {avg_age} лет.')
```

**Задача 5 (со звёздочкой).** Оберните предыдущий код в функцию, которая принимает на вход список и рассчитывает средний возраст, исключая из расчётов выбросы

```python
def middle_age_adv(lst):
    try:
        q1 = np.percentile(lst,25)
        q3 = np.percentile(lst,75)
        iqr = q3-q1
        lower_border = q1-1.5*iqr
        upper_border = q3+1.5*iqr
        sum_age = 0
        cnt_people = 0
        for i in lst:
            if i < upper_border and i > lower_border:
                sum_age +=i
                cnt_people +=1
            avg_age = round(sum_age/cnt_people,0)
        print (f'Средний возраст пользователя {avg_age} лет.')
        return round(avg_age,0)
    except: 
        (f'Список должен содержать только числовые значения и не быть пустым')
```

```python
middle_age_adv(age_lst)
```
